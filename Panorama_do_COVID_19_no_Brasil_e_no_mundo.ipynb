# Panorama do COVID-19 no Brasil e no mundo

<center><img width="50%" src="https://github.com/gabrielrflopes/covid_project/blob/main/capa_cvd_git.png?raw=true"></center>

A COVID-19 é uma doença infecciosa causada pelo coronavírus SARS-CoV-2 que foi inicialmente identificada em dezembro de 2019 na cidade de Wuhan, na China. Desde então, a doença se espalhou rapidamente pelo mundo, levando a uma pandemia global que afetou a vida de milhões de pessoas em todos os continentes.

A COVID-19 é altamente contagiosa e pode ser transmitida de pessoa para pessoa através de gotículas respiratórias expelidas quando uma pessoa infectada tosse, espirra, fala ou respira. 

Os sintomas da doença variam de leves a graves, e incluem febre, tosse seca, fadiga, falta de ar, dores de cabeça e perda do olfato ou do paladar. Em casos graves, a COVID-19 pode levar à pneumonia, falência múltipla de órgãos e morte.

Desde o início da pandemia, milhões de casos foram registrados em todo o mundo, com milhões de mortes causadas pela doença. A pandemia afetou profundamente a vida cotidiana, causando restrições em viagens, comércio, educação e outros setores. 

A resposta à COVID-19 tem sido uma das maiores mobilizações globais de saúde pública na história, com governos, organizações de saúde e empresas trabalhando juntos para conter a propagação da doença, desenvolver tratamentos eficazes e encontrar uma vacina segura e eficaz.

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos disponíveis sobre a doença.

### Dicionário de Variáveis

O dicionário de variáveis abaixo reúne os principais atributos relevantes à análise e aos objetivos propostos para o projeto. As variáveis foram divididas em grupos para facilitar a disposição da informação.

Uma descrição mais completa de todas as variáveis pode ser encontrada no link abaixo:

<h4>Variáveis relacionadas com tempo e localização geográfica</h4>

- `iso_code` - O ISO 3166-1 é parte da norma ISO 3166 que sugere códigos para os nomes de países e dependências.
- `continent` - Continente da localização geográfica.
- `location` - Localização geográfica.
- `date` - Data da observação. 

<h4>Variáveis relacionadas com casos confirmados</h4>

- `total_cases` - Total de casos confirmados de COVID-19 até o dia mais recente.
- `new_cases` - Novos casos diários confirmados de COVID-19. **Importante**: em locais que reportam uma variação diária negativa, a métrica é definida como NaN.
- `total_cases_per_million` - Total de casos confirmados de COVID-19 por milhão de habitantes.
- `new_cases_per_million` - Novos casos diários confirmados de COVID-19 por milhão de habitantes.
- `new_cases_smoothed` - Novos casos confirmados de COVID-19 suavizados numa média de 7 dias.
- `new_cases_smoothed_per_million` - Novos casos confirmados de COVID-19 por milhão de habitantes, suavizados numa média de 7 dias.

<h4>Variáveis relacionadas com mortes por COVID</h4>

- `total_deaths` - Total de mortes atribuídas ao COVID-19.
- `new_deaths` - Novas mortes atribuídas à COVID-19. **Importante**: em locais que reportam uma variação diária negativa, a métrica é definida como NaN.
- `total_deaths_per_million` -Total de mortes atribuídas ao COVID-19 por milhão de habitantes.
- `new_deaths_per_million` -Novas mortes atribuídas ao COVID-19 por milhão de habitantes.
- `new_deaths_smoothed` - Novas mortes atribuídas à COVID-19 suavizadas numa média de 7 dias
- `new_deaths_smoothed_per_million` - Novas mortes atribuídas ao COVID-19 por milhão de habitantes, suavizadas numa média de 7 dias

<h4>Variáveis relacionadas com vacinação e transmissão</h4>

- `reproduction_rate` - Estimativa em tempo-real da taxa de reprodução efetiva (R) da COVID-19. *Fonte: Arroyo Marioli et al. (2020)*.
- `total_vaccinations` - Número total de doses de vacina de COVID-19 administradas.
- `people_vaccinated` - Número total de pessoas que receberam pelo menos uma dose de vacina.
- `people_fully_vaccinated` - Número total de pessoas que receberam todas as doses de vacina indicadas pelo protocolo de vacinação inicial.
- `total_vaccinations_per_hundred` - Número total de doses de vacina de COVID-19 administradas a cada 100 pessoas da população.
- `people_fully_vaccinated_per_hundred` - Número total de pessoas que receberam todas as doses da vacina, prescritas pelo protocolo de vacinação, para cada 100 habitantes da população.

<h4>Variáveis relacionadas com população</h4>

- `population_density` - Número de pessoas dividido pela área territorial, medido em quilômetros quadrados, do ano mais recente.
- `median_age` - Idade mediana da população (Estimativa da ONU para 2020).
- `aged_65_older` - Parcela da população com 65 anos ou mais, do ano mais recente.
- `aged_70_older` - Parcela da população com 70 anos ou mais, do ano mais recente.
- `gdp_per_capita` - PIB per capita do ano mais recente.
- `extreme_poverty` - Parcela da população vivendo na extrema pobreza, ano mais recente desde 2010.
- `excess_mortality` - Diferença de porcentagem entre o número de mortes relatadas semanal ou mensalmente nos anos de 2020-2021, e as projeções para o mesmo período feitas em anos anteriores.

<h4>Variáveis relacionadas ao excesso de mortalidade</h4>

- `excess_mortality` - Diferença em porcentagem entre o número acumulado de mortes em 2020-2021, e o valor projetado para o mesmo período nos anos anteriores.
- `excess_mortality_cumulative` - Diferença em porcentagem entre o número acumulado desde 1 de janeiro de 2020 e as mortes acumuladas projetadas para o mesmo período nos anos anteriores.
- `excess_mortality_cumulative_absolute` - Diferença cumulativa entre as mortes relatadas desde 1 de janeiro de 2020 e o número projetado de mortes para o mesmo período nos anos anteriores. 
- `excess_mortality_cumulative_per_million` - Diferença cumulativa entre as mortes relatadas desde 1 de janeiro de 2020 e o número projetado de mortes para o mesmo período nos anos anteriores, por milhão de habitantes.

## Panorama da COVID-19 no mundo



Iremos iniciar a análise dos dados observando como se deu a pandemia no contexto mundial. 

> Focaremos principalmente em como a doença evolui através do tempo e quais países foram mais atingidos, com relação a número de casos e mortes.

Para começar, vamos focar nos dados mais recentes de número totais de mortes e casos, para definir quais países atualmente foram mais afetados

A data mais recente contida no *dataset* é:

```
data_recente = df_clean.date.max().strftime('%d/%m/%Y')

print(f'A data mais recente é {data_recente}')
```

### **Os países com os maiores números de casos**

Vamos criar um DataFrame paralelo que irá carregar os valores das variáveis filtrados para a data mais recente. 

Para isso, vamos utilizar o método `.loc[]` e selecionar, a partir da coluna `date`, o último dia de atualizações dos dados e, na coluna `continent`, queremos apenas os valores não-nulos.

```
# criando um dataframe para os dados mais recentes, excluindo os valores nulos na coluna continent
df_hoje = df_clean.loc[(df_clean.date == '2023-04-12') & 
                       ~df_clean.continent.isna()]
```

Para organizar os valores segundo nossas variáveis de interesse, usaremos o `.sort_values()` como segue:

```
# ordenar os valores segundo o número total de casos
df_hoje.sort_values(by = 'total_cases', ascending = False)[['iso_code', 'continent', 'location', 'total_cases']].head()
```

```
fig, ax = plt.subplots()

cases = df_hoje.sort_values(by = 'total_cases', ascending = False)[['location', 'total_cases']][4::-1]

cases.plot(x = 'location', 
           y = 'total_cases', 
           kind = 'barh', 
           ax = ax,
           color = 'darkred',
           label = 'Total de casos\n12 de Abril de 2023')

# manipulando o objeto Axes
ax.set_xlabel('Número de casos (em centenas de milhões)', fontweight = 550, labelpad = 15)
ax.set_ylabel('Países', fontweight = 550, labelpad = 15)
ax.ticklabel_format(axis = 'x', useMathText = True)
# ax.set_xlim(0.3e8, )
ax.grid(True, animated = True)
ax.legend()

fig.suptitle('TOP 5 países com maior número de casos confirmados', fontweight = 600, fontsize = 14)
fig.tight_layout()
fig.show()

plt.savefig('top-5-casos.png')
```

Com o gráfico acima, fica fácil visualizar o número de casos nos 5 países mais atingidos. Os EUA lidera com mais de 100 milhões de casos confirmados acumulados, seguido da China, país onde o vírus começou a se espalhar.

A Índia, apesar de sua população de bilhões, mantém a terceira posição, seguida de dois países europeus, a França e a Alemanha. Proporcionalmente ao território e à população, a França e a Alemanha tiveram mais casos por milhão de habitantes. 

```
pct_cases = (df_hoje.total_cases.sort_values(ascending = False)[:5].sum() / df_clean.total_cases.max() * 100).round(2)

print(f'A soma dos casos confirmados no top 5 países corresponde a {pct_cases}% dos casos mundiais.')
```

#### Resumo da seção

- Os 5 países com mais casos acumulados somam mais de 300 milhões de pessoas que ficaram doentes em algum ponto;
- Os EUA lideram com mais de 100 milhões de casos acumulados confirmados, seguido da China;
- A soma dos casos nesses 5 países somam 42,5% dos casos mundiais.

### **Os países com mais mortes atribuídas ao COVID**

### Análise dos dados sobre vacinação

Nesta seção, vamos analisar mais a fundo como se deu a vacinação no mundo.

Para isso, vamos avaliar:

- Os percentis de pessoas que tomaram todas as doses das vacinas;
- A quantidade de pessoas que esses percentis representam;

- **Definindo os percentis**

```
df_clean['people_fully_vaccinated'].describe(percentiles = [.9, .95, .99, .995])
```

```
# Avaliando os percentis de pessoas completamente vacinadas
percentile_90 = df_clean['people_fully_vaccinated'].describe(percentiles = [.25, .5, .75, .9])[7:8][0]
percentile_95 = df_clean['people_fully_vaccinated'].describe(percentiles = [.25, .5, .75, .95])[7:8][0]
percentile_99 = df_clean['people_fully_vaccinated'].describe(percentiles = [.25, .5, .75, .99])[7:8][0]
percentile_995 = df_clean['people_fully_vaccinated'].describe(percentiles = [.25, .5, .75, .995])[7:8][0]
percentile_100 = df_clean['people_fully_vaccinated'].describe(percentiles = [.25, .5, .75])[7:8][0]

print('Percentil de 90% de pessoas completamente vacinada: {:.1e}'.format(percentile_90),
      '\n\nPercentil de 95% de pessoas completamente vacinada: {:.1e}'.format(percentile_95),
      '\n\nPercentil de 99% de pessoas completamente vacinadas: {:.1e}'.format(percentile_99),
      '\n\nPercentil de 99,5% de pessoas completamente vacinadas: {:.1e}'.format(percentile_995),
      '\n\nValor máximo de pessoas completamente vacinadas: {:.1e}'.format(percentile_100))
```

- **Calculando o percentual de pessoas vacinadas**

```
# com os dados sobre os percentis, vamos calcular o quanto representa a população completamente vacinada com relaçaõ à população mundial
pct_wrl_population_90 = percentile_90 / df.population.max() * 100
pct_wrl_population_95 = percentile_95 / df.population.max() * 100
pct_wrl_population_99 = percentile_99 / df.population.max() * 100
pct_wrl_population_995 = percentile_995 / df.population.max() * 100
pct_wrl_population_max = percentile_100 / df.population.max() * 100
```

- **Identificando as datas em que cada percentil foi atingido**

```
# analisando a data com a qual foi atingida essa proporção de vacinações mundiais
wrl_fully_vacc_90 = df_clean['date'].loc[(df_clean.location == 'World') & (df_clean['people_fully_vaccinated'] >= percentile_90)].iloc[0]
wrl_fully_vacc_95 = df_clean['date'].loc[(df_clean.location == 'World') & (df_clean['people_fully_vaccinated'] >= percentile_95)].iloc[0]
wrl_fully_vacc_99 = df_clean['date'].loc[(df_clean.location == 'World') & (df_clean['people_fully_vaccinated'] >= percentile_99)].iloc[0]
wrl_fully_vacc_995 = df_clean['date'].loc[(df_clean.location == 'World') & (df_clean['people_fully_vaccinated'] >= percentile_995)].iloc[0]
wrl_fully_vacc_100 = df_clean['date'].loc[(df_clean.location == 'World') & (df_clean['people_fully_vaccinated'] >= percentile_100)].iloc[0]


# convertendo as datas para o formato string
data_90 = wrl_fully_vacc_90.strftime('%d/%m/%Y')
data_95 = wrl_fully_vacc_95.strftime('%d/%m/%Y')
data_99 = wrl_fully_vacc_99.strftime('%d/%m/%Y')
data_995 = wrl_fully_vacc_995.strftime('%d/%m/%Y')
data_100 = wrl_fully_vacc_100.strftime('%d/%m/%Y')

print(f'No dia {data_90}, {pct_wrl_population_90:.1f}% da população mundial estava vacinada com todas as doses',
      f'\n\nNo dia {data_95}, {pct_wrl_population_95:.1f}% da população mundial estava vacinada com todas as doses',
      f'\n\nNo dia {data_99}, {pct_wrl_population_99:.1f}% da população mundial estava vacinada com todas as doses',
      f'\n\nNo dia {data_995}, {pct_wrl_population_995:.1f}% da população mundial estava vacinada com todas as doses',
      f'\n\nNo dia {data_100}, {pct_wrl_population_max:.1f}% da população mundial estava vacinada com todas as doses')
```

- **Criando um DataFrame para armazenar e visualizar essas informações**

```
dict = {'Percentis': ['90%', '95%', '99%', '99.5%', '100%'], 
        'Datas': [data_90, data_95, data_99, data_995, data_100],
        '% da população': [pct_wrl_population_90, pct_wrl_population_95, pct_wrl_population_99, pct_wrl_population_995, pct_wrl_population_max]
        }

vacc_info = pd.DataFrame(dict)
vacc_info.set_index('Datas', inplace = True)

vacc_info
```

```
fig, ax = plt.subplots(figsize = (6, 5))

vacc_info['% da população'].plot(kind = 'bar', color = 'darkblue', label = 'Vacinação total')

ax.set_xlabel('Data', fontweight = 550, fontsize = 12, labelpad = 10)
ax.set_ylabel('População mundial (%)', fontweight = 550, fontsize = 12, labelpad = 10)
ax.set_title('Evolução da vacinação no mundo', fontweight = 600, fontsize = 14, pad = 15)
ax.legend()

fig.tight_layout()
fig.show()

plt.savefig('vacc_wrl.png')
```
